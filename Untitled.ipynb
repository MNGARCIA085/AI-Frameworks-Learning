```python
import networkx as nx
import matplotlib.pyplot as plt

# Create the graph
g = nx.DiGraph()

g.add_edges_from([
    ("Classification?", "Binary"),
    ("Classification?", "Multiclass"),

    ("Binary", "Logits (raw scores)"),
    ("Binary", "Probabilities (sigmoid)"),

    ("Logits (raw scores)", "BCEWithLogitsLoss"),
    ("Probabilities (sigmoid)", "BCELoss"),

    ("Multiclass", "Sparse Labels"),
    ("Multiclass", "One-hot Labels"),

    ("Sparse Labels", "Logits (raw scores)"),
    ("Sparse Labels", "Probabilities (softmax)"),

    ("One-hot Labels", "Logits (raw scores)"),
    ("One-hot Labels", "Probabilities (softmax)"),

    ("Logits (raw scores)", "CrossEntropyLoss"),
    ("Probabilities (softmax)", "NLLLoss")
])

# Plot the graph
plt.figure(figsize=(10, 6))
pos = nx.spring_layout(g)
nx.draw(g, pos, with_labels=True, node_size=3000, node_color='#a1d0ff', font_size=9, font_weight='bold', arrows=True)
plt.title("Loss Function Selection for Classification")
plt.show()
```

```python
# Create the graph
g = nx.DiGraph()

g.add_edges_from([
    ("Classification?", "Binary"),

    ("Binary", "Raw Outputs"),
    ("Binary", "Probability Outputs"),

    ("Raw Outputs", "Logits (raw scores)"),
    ("Probability Outputs", "Probabilities (sigmoid)"),

    ("Logits (raw scores)", "Balanced Classes"),
    ("Logits (raw scores)", "Imbalanced Classes"),
    ("Probabilities (sigmoid)", "Balanced Classes"),
    ("Probabilities (sigmoid)", "Imbalanced Classes"),

    ("Balanced Classes", "BCEWithLogitsLoss"),
    ("Imbalanced Classes", "BCEWithLogitsLoss + pos_weight"),
    ("Balanced Classes", "BCELoss"),
    ("Imbalanced Classes", "BCELoss + pos_weight")
])

# Print nodes by levels
levels = {
    "Classification?": 0,
    "Binary": 1,
    "Raw Outputs": 2, "Probability Outputs": 2,
    "Logits (raw scores)": 3, "Probabilities (sigmoid)": 3,
    "Balanced Classes": 4, "Imbalanced Classes": 4,
    "BCEWithLogitsLoss": 5, "BCEWithLogitsLoss + pos_weight": 5,
    "BCELoss": 5, "BCELoss + pos_weight": 5
}

nx.set_node_attributes(g, levels, "subset")

print("\nBinary Classification Loss Function Tree (by levels):")
for level in sorted(set(levels.values())):
    print(f"Level {level}: {[node for node in levels if levels[node] == level]}")

# Plot the graph
plt.figure(figsize=(10, 7))
pos = nx.multipartite_layout(g)
nx.draw(g, pos, with_labels=True, node_size=3000, node_color='#a1d0ff', font_size=9, font_weight='bold', arrows=True)
plt.title("Loss Function Selection for Binary Classification")
plt.show()
```
